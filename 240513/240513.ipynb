```python
from Bio import SeqIO

fasta_file = "test.fasta"
sequences = SeqIO.parse(fasta_file, "fasta")

data = dict()

for sequence in sequences:
    for base in sequence.seq:
        if base not in data:
            data[base] = 0
        data[base] += 1

import pandas as pd
df_data = {"base": data.keys(), "base_count": data.values()}
print(df_data)
df = pd.DataFrame(df_data)
df.plot(
    kind="bar",
    x="base",
    y="base_count",
    title="Base count",
    xlabel="Base",
    ylabel="Count",
)
```

```python
from Bio import SeqIO

fasta_file = "test2.fasta"
sequences = SeqIO.parse(fasta_file, "fasta")

data_base = list()

for sequence in sequences:
    data = dict()
    for base in sequence.seq:
        if base not in data:
            data[base] = 0
        data[base] += 1
    data_base.append(data)

import pandas as pd
df_data = {"base": data_base[0].keys()}

for i, data in enumerate(data_base):
    df_data[f"base_count_{i+1}"] = data.values()

print(df_data)

df = pd.DataFrame(df_data)
df.plot(
    kind="bar",
    x="base",
    title="Base count",
    xlabel="Base",
    ylabel="Count",
    figsize=(10, 5),
)
```
